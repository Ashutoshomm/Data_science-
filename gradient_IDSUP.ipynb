```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3])
y=np.array([1,3,2])
plt.scatter(x,y)
plt.xlabel(" x axis")
plt.ylabel(' y axis')
plt.title('Graph')
```

```python
slop_current=0
intercept_rate=0
learning_rate=0.01
iteration=1000
for i in range(iteration):
    y_predicted=slop_current*x +intercept_rate #rms
    slope_grad=-(2/3)*sum(x*(y-y_predicted)) #partial derivative w.r.t m (df/dm)
    intercept_grad=-(2/3)*sum(y-y_predicted)#partialderivative w.r.t c (df/dc)
    slop_current=slop_current-learning_rate*slope_grad
    intercept_rate=intercept_rate-learning_rate*intercept_grad
   # print("m {} ,b {} , rms {} iteration {} ".format(slop_current,intercept_rate,y_predicted,i))
    plt.plot(x,y_predicted,)
    


plt.scatter(x,y)
plt.plot(x,y_predicted,color="red")
    
    
```
